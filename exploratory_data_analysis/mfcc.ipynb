```python
def plot_mfcc(feature, sr):
    plt.figure(figsize=(25, 5))
    librosa.display.specshow(feature, 
             sr=sr,
             x_axis='time')
    plt.colorbar(format='%+2.f')
    plt.show()
```

```python
def extract_mfcc(signal, sr, n_mfcc=13):
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc, n_fft=1024)
    delta_mfcc = librosa.feature.delta(mfcc)
    delta2_mfcc = librosa.feature.delta(delta_mfcc)
    return mfcc, delta_mfcc, delta2_mfcc
```

```python
angry_mfccs, angry_delta, angry_delta2 = extract_mfcc(angry, sr)
happy_mfccs, happy_delta, happy_delta2 = extract_mfcc(happy, sr)
```

```python
plot_mfcc(angry_delta, sr)
plot_mfcc(happy_delta, sr)
```
