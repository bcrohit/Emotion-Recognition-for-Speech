```python
import librosa
import librosa.display
import numpy as np

import utilityFunctions as utf
import matplotlib.pyplot as plt
```

```python
dirpath = r"D:\SER-RO-MAHA\Sample Speech Data"

neutral, sr = utf.return_audio_array('03-01-01-01-01-01-01.wav', dirpath, 'Neutral')
happy, _ = utf.return_audio_array('03-01-03-01-01-01-01.wav', dirpath, 'Happy')
sad, _ = utf.return_audio_array('03-01-04-01-01-01-01.wav', dirpath, 'Sad')
angry, _ = utf.return_audio_array('03-01-05-01-01-01-01.wav', dirpath, 'Anger')
fear, _ = utf.return_audio_array('03-01-06-01-01-01-01.wav', dirpath, 'Fear')
disgust, _ = utf.return_audio_array('03-01-07-01-01-01-01.wav', dirpath, 'Disgust')
surprise, _ = utf.return_audio_array('03-01-08-01-01-01-01.wav', dirpath, 'Surprise')
```

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512
```

```python
def plot(feature, hop_length, frame_size, name):
    frames = range(len(feature))
    t = librosa.frames_to_time(frames, hop_length=hop_length)
    
    plt.figure(figsize=(25, 10))
    plt.plot(t, feature, color='r')
    plt.title(name)
```

```python
def extract_spectral_centroid(signal, frame_size, hop_length):  
    sc_signal = librosa.feature.spectral_centroid(y=signal, sr=sr, n_fft=frame_size, hop_length=hop_length)[0]
    return sc_signal
    
def extract_spectral_bandwidth(signal, frame_size, hop_length):
    bw_signal = librosa.feature.spectral_bandwidth(y=signal, sr=sr, n_fft=frame_size, hop_length=hop_length)[0]
    return bw_signal
```

```python
sc_angry = extract_spectral_centroid(angry, FRAME_SIZE, HOP_LENGTH)
sc_neutral = extract_spectral_centroid(neutral, FRAME_SIZE, HOP_LENGTH)
sc_happy = extract_spectral_centroid(happy, FRAME_SIZE, HOP_LENGTH)

plot(sc_angry, HOP_LENGTH, FRAME_SIZE, '')
plot(sc_neutral, HOP_LENGTH, FRAME_SIZE, 'Neutral')
plot(sc_happy, HOP_LENGTH, FRAME_SIZE, 'Happy')
```

```python
bw_angry = extract_spectral_bandwidth(angry, FRAME_SIZE, HOP_LENGTH)
bw_neutral = extract_spectral_bandwidth(neutral, FRAME_SIZE, HOP_LENGTH)
bw_happy = extract_spectral_bandwidth(happy, FRAME_SIZE, HOP_LENGTH)

plot(bw_angry, HOP_LENGTH, FRAME_SIZE, '')
plot(bw_neutral, HOP_LENGTH, FRAME_SIZE, 'Neutral')
plot(bw_happy, HOP_LENGTH, FRAME_SIZE, 'Happy')
```
